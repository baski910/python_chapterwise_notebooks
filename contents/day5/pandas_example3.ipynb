```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import dateutil
```

```python
data=sm.datasets.co2.load_pandas()
df=data.data
```

```python
print(df.head())
```

```python
df1 = df['co2'].resample('MS').mean()  # MS - Month Start   M - Month end
```

```python
print(df1.head())
```

```python
# fill NaN values
df1=df1.fillna(df1.bfill()) 
```

```python
print(df1.head())
```

```python
df1.plot()
plt.show()
```
