```python
import csv # used for reading csv files

def csv_reader(fobj):
    records = csv.reader(fobj)
    #print(records)
    for record in records:
        #print(record)
        print(" ".join(record))
```

```python
with(open('uk-500.csv','r')) as fh:
    csv_reader(fh)
```

```python

def csv_dict_reader(fobj):
    records = csv.DictReader(fobj,delimiter=',')
    #print(records)
    for record in records:
        #print(record)
        print(record['first_name'],record['last_name'])
```

```python
with(open('uk-500.csv','r')) as fh:
    csv_dict_reader(fh)
```
