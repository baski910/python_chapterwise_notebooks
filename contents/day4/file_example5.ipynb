```python
import re
fh = open('readme.txt','r')

fh.read() # by this we reach EOF 

fh.seek(0,0)  # resets the cursor to the beginning of the file

for line in fh.readlines():
    v = re.search(r'hello',line)
    if v:
        continue
    print(line)

fh.close()
```
