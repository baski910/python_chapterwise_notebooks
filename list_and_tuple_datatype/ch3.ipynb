<h1>List Functions</h1>
<h2>Function to add elements to list</h2>
<ul>
    <li>append - takes 1 argument. argument is added to the end of the list. <br>
        increases the size of the list by 1.</li>
    <li>extend - takes 1 argument. argument must be a sequence. adds the elements of<br>
        sequence to end of the list. Increases the size of the list by size of argument passed</li>
    <li>insert - takes 2 argument. 1st parameter is position. 2nd parameter is the<br>
        element to be added. add the 2nd parameter at the position specified by 1st parameter<br>
        and increase the size of the list by 1</li>
</ul>

```python
"""Examples for adding elements to list"""

list1 = [10,20,30,40]

print("elements of list1:",list1)
print("size of list1:",len(list1))

list1.append(50)

print("#"*20+" after append "+"#"*20)

print("elements of list1:",list1)
print(len(list1))


list1 = [10,20,30,40]

print("elements of list1:",list1)
print("size of list1:",len(list1))

list1.append([50,60,70]) # append a sequence

print("#"*20+" after append "+"#"*20)

print("elements of list1:",list1)
print(len(list1))

list1 = [10,20,30,40]

print("elements of list1:",list1)
print("size of list1:",len(list1))

list1.extend([50,60,70])

print("#"*20+" after extend "+"#"*20)

print("elements of list1:",list1)
print(len(list1))



list1 = [10,20,30,40]

print("elements of list1:",list1)
print("size of list1:",len(list1))

list1.insert(2,50)

print("#"*20+" after insert "+"#"*20)

print("elements of list1:",list1)
print(len(list1))


list1 = [10,20,30,40]

print("elements of list1:",list1)
print("size of list1:",len(list1))

list1.insert(2,[50,60,70]) # insert a sequence

print("#"*20+" after insert "+"#"*20)

print("elements of list1:",list1)
print(len(list1))
```
