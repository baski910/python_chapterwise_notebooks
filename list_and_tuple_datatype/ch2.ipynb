<h1>List References</h1>
<p>List are mutable objects. References are not modified on modifying list elements</p>
<p>Creating a list can depicted as below</p>
<img src="assigning_list1.jpg" width="100" height="200">
<p>Assigning a list can depicted as below</p>
<img src="assigning_list_reference1.jpg" width="100" height="300">
<p>In essence, both list1 and list2 are pointing to the same object and have the same reference</p>

```python
"""list as mutable objects"""
list1 = [10,20,30,40,50]  # the list elements can be of any data type. need not be of same datatype

print("reference of list1:",id(list1))
print("elements of list1:",list1)

list1[1] = 100
print("#"*20+"after modification"+"#"*20)

print("reference of list1:",id(list1))
print("elements of list1:",list1)
```

```python
"""Examples for list references"""
list1 = [10,20,30,40,50]  # the list elements can be of any data type. need not be of same datatype

# the below assignment will assign the reference of list1 to list2
# changes made to list1 or list2 will be reflected in the both the lists

list2 = list1

print("reference of list1:",id(list1))
print("reference of list2:",id(list2))

print("elements of list1:",list1)
print("elements of list2:",list2)

list2[2]=80

print("#"*20+"after modification"+"#"*20)

print("elements of list1:",list1)
print("elements of list2:",list2)
```

<h1>Copying lists</h1>
<ul>
    <li>Shallow Copy</li>
    <li>Deep Copy</li>
</ul>
<p>In shallow copy, the lists get different references however, the inner elements have the same reference</p>
<p>In case of shallow copy, the modification made to one list may affect other list depending on the elements<br>
    being immutable or mutable object. If the element is an immutable object, the changes will not be reflected,<br>
otherwise it will be reflected</p>


<p>
    <img src="shallow_copy_immutable1.jpg" width="200" height="200">
    <img src="shallow_copy_mutable1.jpg" width="200" height="200">
</p>

<p>In deep copy, the lists and their inner elements get different refernces. Hence the modification in one list will<br>
    not affect the other</p>
<p>
    <img src="deepcopy_1.jpg" width="200" height="200">
</p>
    
 

```python
"""Examples for shallow copy"""

# list can be copied by using list function 'copy'
# list1 = ['a','b','c','d','e','f']
# list2 = list1.copy()


list1 = ['a','b','c','d','e','f'] # list elements are immutable objects

print("reference of list1:",id(list1))

list2 = list1[:] # list2 will be created with new reference but with set of elements in list1

print("reference of list2:",id(list2))

print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

print("elements of list1:",list1)

print("elements of list2:",list2)

list2[1] = 'g' # changing element 'b' to 'g' in list2

print("#"*20+" after modification "+"#"*20)

print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

# modification made to list2 does not affect list1
print("elements of list1:",list1)

print("elements of list2:",list2)
```

```python
"""Examples for shallow copy"""

list1 = [['a','b'],['c','d'],['e','f']] # list elements are mutable objects

print("reference of list1:",id(list1))

list2 = list1[:] # list2 will be created with new reference but with set of elements in list1

print("reference of list2:",id(list2))

print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

print("elements of list1:",list1)

print("elements of list2:",list2)

list2[1][1] = 'g' # changing element 'd' to 'g' in list2

print("#"*20+" after modification "+"#"*20)
print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

# modification made to list2 reflects in list1
print("elements of list1:",list1)

print("elements of list2:",list2)
```

```python
"""Examples for deep copy"""
from copy import deepcopy

list1 = [['a','b'],['c','d'],['e','f']] # list elements are mutable objects

print("reference of list1:",id(list1))

list2 = deepcopy(list1) # list2 will be created with new reference but with set of elements in list1

print("reference of list2:",id(list2))

print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

print("elements of list1:",list1)

print("elements of list2:",list2)

list2[1][1] = 'g' # changing element 'b' to 'g' in list2

print("#"*20+" after modification "+"#"*20)
print("reference of element at index 0 of list1:",id(list1[0]))
print("reference of element at index 0 of list2:",id(list2[0]))

# modification made to list2 reflects in list1
print("elements of list1:",list1)

print("elements of list2:",list2)
```
