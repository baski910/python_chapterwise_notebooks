<h1>Strings</h1>
<h2>Strings as sequence</h2>
<p>Strings are immutable objects</p>
<p>In python, strings are treated as continguous set of characters. They are treated as<br>
    array of characters.</p>
<p>The characters of strings can be accessed using index starting from 0 or -1.<br>
    Positive index works from the beginning of the string and negative index works from end of the string</p>
<p>Strings can be iterated</p>
<p>A substring can be formed by using slicing operator ':'. It takes start index, end index and returns the<br>
    character(s) in between</p>
<p>Strings can be concatenated using '+' operator and multiplied using '*' operator</p>

```python
"""Examples using string"""

s1="helloworld"

#iterate over each character
for char in s1:
    print(char)
    
#return character at index 4
print(s1[4]) # output will be 'o'

#preturn the last character using negative index
print(s1[-1]) # output will be 'd'

#retuns character starting from index 1 upto index 5(not including). i.e., prints character at index 1,2,3 and 4
print(s1[1:5])

#returns characters from beginning of the string upto index specified by upper limit
print(s1[:5]) # output will be 'hello'

#returns characters upto end of the string from index specified by lower limit
print(s1[5:]) # output will be 'world'

#returns a new string with same set of characters
print(s1[:]) # output will be 'helloworld'

"""extended slicing can be done on a string as follows:
string[start:stop:step]
characters within start and stop index jumping in steps 
"""

print(s1[1:8:2]) # output will be 'elwr' - i.e., characters at index 1,3,5 and 7

#a string can be reversed using extended slicing
print(s1[::-1]) # output will be 'dlrowolleh'

"""concatenate strings"""
s1="hello"
s2="world"

print(s1+s2) # output will be 'helloworld'

"""string multiplication"""
s1='#'

print(s1*60) # output will be '#' printed for 60 times
```

<h1>String Functions</h1>
<ul>
    <li>upper - returns a string with all UPPERCASE</li>
    <li>find - returns the index of 1st occurance of substring inside the string. if substring is not found returns -1</li>
    <li>index - returns the index of 1st occurance of substring inside the string. if substring is not found throws error</li>
    <li>count - returns the number of occcurances of substring inside the string. returns 0 if substring not found</li>
    <li>isdigit - returns true if the character is a digit</li>
    <li>split - returns a list, by spliting the string using the delimiter passed as an argument. whitespace is the default delimiter</li>
</ul>

```python
"""Examples for string functions"""

s1="helloworld"

print(s1.upper()) # output will be HELLOWORLD

print(s1.find('w')) # output will be 5

print(s1.find('a')) # output will be -1

print(s1.index('w')) # output will be 5

print(s1.index('a')) # throws error

print(s1.count('l')) # output will be 3

print(s1.count('a')) # output will be 0

s1="cat sat on the mat"

print(s1.split()) # output will be ['cat','sat','on','the','mat']

s1="2020-03-15"

print(s1.split('-')) # output will be ['2020','03','15']
```

<h2>Using '%' operator</h2>
<p>String formating using' %' operator</p>
<p>If format requires a single argument, values may be a single non-tuple object.</p>
<p>If format requires multiple argument values must be a tuple with exactly the number of<br>  
    items specified by the format string, or a single mapping object (for example, a dictionary).</p>
<h2>Using 'format' function</h2>
<p>The .format() method of the str type is an extremely convenient way to format text exactly the way you want it</p>
<p>Here is the general form:</p>
<p>template.format(p0, p1, ..., k0=v0, k1=v1, ...)</p>
<p>The template is a string containing a mixture of one or more format codes embedded in constant text.</p>
<p>The format method uses its arguments to substitute an appropriate value for each format code in the template</p>

```python
"""Examples for using '%' operator"""
"""
%s replaces strings
%f replaces float
%d replace integer
"""
# to display the value with specified number of decimal precision, '%' operator is used

pi = 3.14159

print("pi=%.2f" %pi) # output will be 'pi=3.14'

# to produce a output something like 'bob has scored 80 marks in maths', 
# using variables name,mark and subject with '%' opertor we do as given below

name='bob'
mark=80
subject='maths'

print("%s has scored %d marks in %s" %(name,mark,subject))

# the output produced in the above statement, can also be done using concatenation 
# However, it involves type casting whereever necessary. 
# Writing this way has less readability when number of variables are more

print(name+" has scored "+str(mark)+" marks in "+subject)

# Even with the information(data) in the form of a dictionary, the same output could be produced

d1 = {'name':'bob','mark':80,'subject':'maths'}

print("%(name)s has scored %(mark)d marks in %(subject)s" %d1
```

```python
"""Examples for using format function"""

# in the below statement the pair of curly braces are substituted with the parameters of 
# format function is same order how they were passed
print("We have {} students in {}".format(60, "Computer Science"))  # output will be - We have 60 students in Computer Science

# if the values passed in different order to the format function, index can be used to substitute appropriately
print("We have {1} students in {0}.".format("Computer Science",60)) # output will be the same as above statement

# variables can also be passed to format function

department='Computer Science'
num_of_students=60
print("We have {1} students in {0}.".format(department,num_of_students)) # output will be the same as above statement

# dictionary values can also be substituted in format function

d1= {'department':'Computer Science','num_of_students':60}

print("We have {num_of_students} students in {department}.".format(**d1)) # output will be the same as above statement
                                                     # ** - refers unknown number of key and value pairs
```
