```python
import pandas as pd
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression


df = pd.read_csv('Weather.csv')

#print(df.head())

x = df['MinTemp'].values.reshape(-1,1)
y = df['MaxTemp'].values.reshape(-1,1)

#x = df['MinTemp'].values.reshape(-1,1)

#print(x)
plt.scatter(x,y)
plt.ylabel('depdendent')
plt.xlabel('independent')
plt.show()
```

```python
# create training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
linreg = LinearRegression()

# call the fit function of LinearRegression for training the data
linreg.fit(x_train,y_train)
```

```python
# using x_test as in input get predicted values
y_pred = linreg.predict(x_test)
```

```python
# create a new datafrme again using y(predicted) and y(actual)

df1 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})

df1 = df1.head(25)

df1.plot(kind='bar')
```

```python
# get the score of our model
print(linreg.score(x_test,y_test))
```
